### While Loop

In Python, a `while` loop is used to repeatedly execute a block of code as long as a specified condition is True. 

The loop continues until the condition becomes False.

```python
# Example: Printing numbers from 1 to 5 using a while loop
a = 1
while a <= 5:
    a += 1
    # a= a+1
    print(a)
```

```python
a = 1 
b = 10
while a <=b :
    a = a +1
    print(a)
```

```python
a = 1
b = 10 
while a <=b :
    print(a)
    a = a *2
else :
    print("print this else block")
```

```python
a = 1 
b = 10 
while a <10 :
    # print(a)
    if a == 5:
        break
    a = a +1
    print(a)
else :
    print("print this ")
```

```python
for i in range(10):
    if i==5:
        break
    print(i)
else:
    print("print this")
```

```python
a = 1 
b = 10 
while a < b :
    print(a)
    if a == 3:
        continue
    a = a +1
```

continue Statement:
The continue statement is used to skip the current iteration of a loop and move to the next iteration.

```python
# Example: Using 'continue' to skip printing even numbers
for num in range(1, 6):
    if num % 2 == 0:
        continue
    print(num)
```

```python
# Example: Using 'pass' to do nothing inside the loop
for num in range(1, 6):
    pass
```

```python
a = 1 
b = 10 
while a < b :
    a = a +1
    if a == 3:
        pass
```

```python
a = 1 
b = 10 
while a < b :
    a = a +1
    if a == 3:
        print("a now is 3")
        pass  
```

```python
# Example: Using 'break' to stop the loop when the value reaches 3
for num in range(1, 6):
    if num == 3:
        break
    print(num)
```
